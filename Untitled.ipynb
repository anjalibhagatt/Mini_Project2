```python
from pystac_client import Client
```

```python
client = Client.open("https://earth-search.aws.element84.com/v1/")
client
```

```python
collections = client.get_collections()
```

```python
for collection in collections:
    print(collection.id)
```

```python
collection=client.get_collection('sentinel-2-l2a')
collection
```

```python
result = client.search(
    collections=["sentinel-2-l2a"]
    
)
```

```python
result.matched()
```

```python
point = {
    "type" : "Point",
    "coordinates":(6.8962, 52.22502)
}
```

```python
result = client.search(
    collections=["sentinel-2-l2a"],
    intersects=point,
    datetime="2023-01-0
    max_items=10
    
)
```

```python
result.matched()
```

```python
items = result.item_collection()
items
```
